```python
Q1. Explain Class and Object with respect to Object-Oriented Programming. Give a suitable example?
Ans1. A class is a user-defined data type that defines a blueprint for creating objects. It contains attributes, which are variables that hold data, and methods, which are functions that can be called to perform operations on the data. Classes allow us to encapsulate data and functionality into a single unit, making it easier to manage and reuse code.
An object, on the other hand, is an instance of a class. It represents a specific entity that has its own data and methods. When an object is created, it is initialized with the attributes and methods defined in its class.
class Car:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year
        self.mileage = 0
    
    def drive(self, miles):
        self.mileage += miles
    
    def get_description(self):
        return f"{self.year} {self.make} {self.model}"
```

```python
Q5. What is inheritance? Give an example for each type of inheritance?
Ans5. There are four types of inheritances
Single Inheritance: In single inheritance, a derived class inherits from a single base class. The derived class has access to all the public and protected members of the base class, but cannot access its private members.
class Animal:
    def __init__(self, name):
        self.name = name
    
    def speak(self):
        print("Animal is speaking.")

class Dog(Animal):
    def __init__(self, name):
        super().__init__(name)
    
    def speak(self):
        print("Woof!")

dog = Dog("Buddy")
print(dog.name)  # Output: "Buddy"
dog.speak()  # Output: "Woof!"

Multiple Inheritance: In multiple inheritance, a derived class inherits from two or more base classes. The derived class has access to all the public and protected members of all the base classes.
class A:
    def a(self):
        print("A")

class B:
    def b(self):
        print("B")

class C(A, B):
    def c(self):
        print("C")

obj = C()
obj.a()  # Output: "A"
obj.b()  # Output: "B"
obj.c()  # Output: "C"

Multi-level Inheritance: In multi-level inheritance, a derived class inherits from a base class, which in turn inherits from another base class. The derived class has access to all the public and protected members of both base classes.
class Animal:
    def speak(self):
        print("Animal is speaking.")

class Dog(Animal):
    def bark(self):
        print("Woof!")

class Labrador(Dog):
    def fetch(self):
        print("Fetching...")

lab = Labrador()
lab.speak()  # Output: "Animal is speaking."
lab.bark()  # Output: "Woof!"
lab.fetch()  # Output: "Fetching..."


Hierarchical Inheritance: In hierarchical inheritance, multiple derived classes inherit from a single base class. The derived classes have access to all the public and protected members of the base class.
class Animal:
    def speak(self):
        print("Animal is speaking.")

class Dog(Animal):
    def bark(self):
        print("Woof!")

class Cat(Animal):
    def meow(self):
        print("Meow!")

dog = Dog()
dog.speak()  # Output: "Animal is speaking."
dog.bark()  # Output: "Woof!"

cat = Cat()
cat.speak()  # Output: "Animal is speaking."
cat.meow()  # Output:
```
